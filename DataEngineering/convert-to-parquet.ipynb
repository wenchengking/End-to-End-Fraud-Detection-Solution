```python
import pandas as pd
df = pd.read_csv('train_transaction.csv')
df.head()
```

```python
df.to_parquet('train_transaction.parquet', engine='auto', compression='snappy')
print('done')
```

```python
import os

file_name = "train_transaction.parquet"
file_stats = os.stat(file_name)

print(file_stats)
print(f'File Size in Bytes is {file_stats.st_size}')
print(f'File Size in MegaBytes is {file_stats.st_size / (1024 * 1024)}')
```
